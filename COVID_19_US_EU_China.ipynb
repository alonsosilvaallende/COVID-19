```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```
known_cases = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/total_cases.csv",\
                          index_col='date')
known_cases = known_cases.fillna(0)
```

```
deaths = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/total_deaths.csv", index_col='date')
deaths = deaths.fillna(0)
```

```
EU_list = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', \
           'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', \
           'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', \
           'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', \
           'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', \
           'Spain', 'Sweden']
```

```
EU_known_cases = known_cases[EU_list]
EU_known_cases = EU_known_cases.sum(axis=1)
EU_deaths = deaths[EU_list]
EU_deaths = EU_deaths.sum(axis=1)
```

```
China_known_cases = known_cases["China"]
China_deaths = deaths["China"]
```

```
US_known_cases = known_cases["United States"]
US_deaths = deaths["United States"]
```

```
# Population in millions from Wikipedia
EU_population = 447
China_population = 1427
US_population = 328
```

```
China_known_cases.tail(3)
```

```
EU_known_cases.tail(3)
```

```
fig, ax = plt.subplots(figsize=(10,7))


plt.plot(EU_known_cases, label='European Union', linewidth=2, marker='o')
plt.plot(China_known_cases, label='China', linewidth=2, marker='o')
plt.plot(US_known_cases, label='United States', linewidth=2, marker='o')

ax.set_title('COVID-19: Number of known cases vs date', fontsize='x-large')
ax.set_xlabel('date', fontsize='x-large')
ax.set_ylabel('number of known cases', fontsize='x-large')
ax.set_xticks(np.arange(0, len(known_cases['China']),5))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```

```
fig, ax = plt.subplots(figsize=(10,7))

plt.plot(EU_known_cases/EU_population, label='European Union', linewidth=2, marker='o')
plt.plot(China_known_cases/China_population, label='China', linewidth=2, marker='o')
plt.plot(US_known_cases/US_population, label='United States', linewidth=2, marker='o')

ax.set_title('COVID-19: Number of known cases per million people vs date', fontsize='x-large')
ax.set_xlabel('date', fontsize='x-large')
ax.set_ylabel('number of known cases per million people', fontsize='x-large')
ax.set_xticks(np.arange(0, len(known_cases['China']),5))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```

```
fig, ax = plt.subplots(figsize=(10,7))

plt.plot(EU_deaths, label='European Union', linewidth=2, marker='o')
plt.plot(China_deaths, label='China', linewidth=2, marker='o')
plt.plot(US_deaths, label='United States', linewidth=2, marker='o')

ax.set_title('COVID-19: Number of deaths vs date', fontsize='x-large')
ax.set_xlabel('date', fontsize='x-large')
ax.set_ylabel('number of deaths', fontsize='x-large')
ax.set_xticks(np.arange(0, len(deaths["China"]), 5))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```

```
fig, ax = plt.subplots(figsize=(10,7))

plt.plot(EU_deaths/EU_population, label='European Union', linewidth=2, marker='o')
plt.plot(China_deaths/China_population, label='China', linewidth=2, marker='o')
plt.plot(US_deaths/US_population, label='United States', linewidth=2, marker='o')

ax.set_title('COVID-19: Number of deaths per million people vs date', fontsize='x-large')
ax.set_xlabel('date', fontsize='x-large')
ax.set_ylabel('number of deaths per million people', fontsize='x-large')
ax.set_xticks(np.arange(0, len(known_cases['China']),5))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```
