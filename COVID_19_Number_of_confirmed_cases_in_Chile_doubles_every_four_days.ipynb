```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use('seaborn')
```

```
total_cases = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/total_cases.csv", index_col='date')
total_cases = total_cases.fillna(method='pad')
```

```
Chile_total_cases = total_cases["Chile"]
Chile_total_cases = Chile_total_cases[Chile_total_cases>=1]
```

```
Chile = pd.Series()
counter = 0
for i, v in Chile_total_cases.items():
  if counter % 4 == 0:
    Chile[i] = v
  counter += 1
```

```
# set width of bar
barWidth = 0.4
 
# set height of bar
bars1 = [1, 2, 10, 46, 150, 684]
bars2 = [1, 5, 23, 75, 342, 746]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
# Make the plot
fig, ax = plt.subplots(figsize=(10,7))
for i, val in enumerate(bars1):
   ax.text(i-0.1, val+10, str(int(val)), fontsize='x-large')
for i, val in enumerate(bars2):
   ax.text(i+0.3, val+10, str(int(val)), fontsize='x-large')

plt.bar(r1, bars1, color='C0', width=barWidth, edgecolor='white', label='Doble con respecto a 4 días antes')
plt.bar(r2, bars2, color='C1', width=barWidth, edgecolor='white', label='Casos confirmados totales')
 
# Add xticks on the middle of the group bars
plt.title('COVID-19: Número de casos confirmados totales cada 4 días desde el primer caso confirmado', fontsize='x-large')
plt.xlabel('fecha (día/mes)', fontsize='x-large')
plt.ylabel('número de casos confirmados totales', fontsize='x-large')
plt.xticks([r + barWidth for r in range(len(bars1))], ['04/03', '08/03', '12/03', '16/03', '20/03', '24/03'], fontsize='x-large')
 
# Create legend & Show graphic
plt.legend(fontsize='x-large')
plt.show()
```
