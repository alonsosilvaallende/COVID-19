```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```
known_cases = pd.read_csv("https://covid.ourworldindata.org/data/total_cases.csv",\
                          index_col='date')
known_cases = known_cases.fillna(0)
```

```
deaths = pd.read_csv("https://covid.ourworldindata.org/data/total_deaths.csv", index_col='date')
deaths = deaths.fillna(0)
```

```
EU_list = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', \
           'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', \
           'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', \
           'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', \
           'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', \
           'Spain', 'Sweden']
```

```
EU_known_cases = known_cases[EU_list]
EU_known_cases = EU_known_cases.sum(axis=1)
EU_deaths = deaths[EU_list]
EU_deaths = EU_deaths.sum(axis=1)
```

```
China_known_cases = known_cases["China"]
China_deaths = deaths["China"]
```

```
US_known_cases = known_cases["United States"]
US_deaths = deaths["United States"]
```

```
Chile_known_cases = known_cases["Chile"].copy()
Chile_deaths = deaths["Chile"]
```

```
# Population in millions from Wikipedia
EU_population = 447
China_population = 1427
US_population = 328
Chile_population = 17
```

```
fig, ax = plt.subplots(figsize=(10,7))

plt.plot(EU_known_cases/EU_population, label='Unión Europea', linewidth=2, marker='o')
plt.plot(China_known_cases/China_population, label='China', linewidth=2, marker='o')
plt.plot(US_known_cases/US_population, label='Estados Unidos', linewidth=2, marker='o')
plt.plot(Chile_known_cases/Chile_population, label='Chile', color = 'C3', linewidth=2, marker='o')

ax.set_title('COVID-19: Número de casos confirmados por millón de habitantes', fontsize='x-large')
ax.set_xlabel('fecha (año-mes-día)', fontsize='x-large')
ax.set_ylabel('número de casos confirmados por millón de habitantes', fontsize='x-large')
ax.set_xticks(np.arange(len(EU_known_cases)))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```

```
fig, ax = plt.subplots(figsize=(10,7))

plt.plot(EU_deaths/EU_population, label='Unión Europea', linewidth=2, marker='o')
plt.plot(China_deaths/China_population, label='China', linewidth=2, marker='o')
plt.plot(US_deaths/US_population, label='Estados Unidos', linewidth=2, marker='o')
plt.plot(Chile_deaths/Chile_population, label='Chile', linewidth=2, marker='o')

ax.set_title('COVID-19: Número de muertes por millón de habitantes', fontsize='x-large')
ax.set_xlabel('fecha (año-mes-día)', fontsize='x-large')
ax.set_ylabel('número de muertes por millón de habitantes', fontsize='x-large')
ax.set_xticks(np.arange(len(EU_known_cases)))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```
