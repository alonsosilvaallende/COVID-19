```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```
known_cases = pd.read_csv("https://covid.ourworldindata.org/data/total_cases.csv",\
                          index_col='date')
known_cases = known_cases.fillna(0)
```

```
deaths = pd.read_csv("https://covid.ourworldindata.org/data/total_deaths.csv", index_col='date')
deaths = deaths.fillna(0)
```

```
countries_list = ['China', 'South Korea', 'Italy', 'Spain', 'France', 'Japan', 'Chile']
```

```
# Population in millions from Wikipedia
population = [1427, 51, 60, 46, 67, 126, 17]
```

```
fig, ax = plt.subplots(figsize=(10,7))

for i, country in enumerate(countries_list):
  plt.plot(known_cases[country]/population[i], label=country, linewidth=2, marker='o')

ax.set_title('COVID-19: Número de casos confirmados por millón de habitantes', fontsize='x-large')
ax.set_xlabel('fecha (año-mes-día)', fontsize='x-large')
ax.set_ylabel('número de casos confirmados por millón de habitantes', fontsize='x-large')
ax.set_xticks(np.arange(len(known_cases['Italy'])))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```

```
fig, ax = plt.subplots(figsize=(10,7))

for i, country in enumerate(countries_list):
  plt.plot(deaths[country]/population[i], label=country, linewidth=2, marker='o')

ax.set_title('COVID-19: Número de muertes por millón de habitantes', fontsize='x-large')
ax.set_xlabel('fecha (año-mes-día)', fontsize='x-large')
ax.set_ylabel('número de muertes por millón de habitantes', fontsize='x-large')
ax.set_xticks(np.arange(len(deaths['Italy'])))
plt.xticks(rotation=90)
plt.legend(fontsize='x-large')
plt.yscale("log")
plt.show()
```
